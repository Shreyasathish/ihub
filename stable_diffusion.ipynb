```python
import torch
from torch import nn
from torch.nn import functional as F
from torchvision import models
from torchvision.transforms import functional as TF
from PIL import Image
import numpy as np

# Load the Stable Diffusion model
model = models.StableDiffusion(
    unet_dim=128,
    text_encoder_dim=128,
    timesteps=1000,
    beta_schedule="linear",
    beta_start=0.0001,
    beta_end=0.02,
    text_encoder_name="clip-ViT-B-32",
)

# Load the CLIP text encoder
text_encoder = models.CLIPTextEncoder("clip-ViT-B-32")

# Define a function to generate an image from a text prompt
def generate_image(prompt, height=512, width=512, num_steps=50):
    # Encode the text prompt
    text_encoding = text_encoder.encode(prompt)

    # Create a random noise tensor
    noise = torch.randn(1, 3, height, width)

    # Iterate over the diffusion steps
    for i in range(num_steps):
        # Compute the noise schedule
        t = torch.tensor([i / num_steps])
        noise_schedule = model.beta_schedule(t)

        # Compute the denoising step
        noise = model.unet(noise, t, text_encoding)

        # Compute the image
        image = model.decode(noise)

    # Return the generated image
    return image

# Test the function
prompt = "A futuristic cityscape with towering skyscrapers and flying cars"
image = generate_image(prompt, height=512, width=512, num_steps=50)

# Display the generated image
Image.fromarray(np.uint8(image.detach().numpy().transpose(1, 2, 0) * 255)).show()
```

```python
# Install the diffusers library first:
# pip install diffusers transformers torch

import torch
from diffusers import StableDiffusionPipeline
from PIL import Image

# Load the Stable Diffusion model pipeline
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")

# Define a function to generate an image from a text prompt
def generate_image(prompt, height=512, width=512, num_inference_steps=50):
    # Generate the image
    with torch.no_grad():
        image = pipe(prompt, height=height, width=width, num_inference_steps=num_inference_steps).images[0]

    # Return the generated image
    return image

# Test the function
prompt = "A futuristic cityscape with towering skyscrapers and flying cars"
image = generate_image(prompt, height=512, width=512, num_inference_steps=50)

# Display the generated image
image.show()
```
